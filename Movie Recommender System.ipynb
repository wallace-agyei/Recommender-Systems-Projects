```python
#Total number of ratings by ussers
total_rating = df.groupby('title')['rating'].count()
```

```python
#we now create a dataframe for average rating and number of ratingss
data = pd.DataFrame()
data['Average_rating'] = average_rating
data['Total_rating'] = total_rating
```

```python
df = df.iloc[0:1500000]
```

```python
#recomending movies that are simiar
movie_title = df.pivot_table(index ='userId',columns='title',values='rating')
```

```python
similar_ratings  =  movie_title['Ace Ventura: When Nature Calls (1995)']
similar_ratings.head(30)
```

```python
simialr_to_others  = movie_title.corrwith(similar_ratings)
corr = pd.DataFrame(simialr_to_others,columns =['Correlation'])
corr.dropna(inplace = True)
```

Movie Recommendation System

```python
def movie_recommendation():
    
    movie = input('Please Enter Movie Name : ')#taking movie as input
    num = int(input('How many recommendation you need : ' )) #number of recommendation 
    user_rating = movie_title[movie]
    similar_movie = movie_title.corrwith(user_rating)
    similar_movie = pd.DataFrame(similar_movie, columns =['Correlation'])
    similar_movie = similar_movie.join(data['Total_rating'])
    recomm = similar_movie.sort_values(by = ['Correlation', 'Total_rating'],ascending=False)
    recommend = pd.DataFrame(recomm)
    print('Following are the {} Recommended Movies'.format(num))
    
    return recommend.head(num)
```

```python
movie_recommendation()
```
